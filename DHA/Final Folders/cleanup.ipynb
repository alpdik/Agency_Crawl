```python
df = pd.read_csv('dha_articles(3).csv')
df_clean = df.dropna(subset=['headline'])
df_clean.to_csv('dha_articles_cleaned(3).csv', index=False)
```

```python
import pandas as pd
import glob

# Find all cleaned CSV files
csv_files = glob.glob('dha_articles_cleaned(*).csv')

# Read and concatenate all DataFrames
dfs = [pd.read_csv(f) for f in csv_files]
merged_df = pd.concat(dfs, ignore_index=True)

# Save to Parquet
merged_df.to_parquet('dha_articles_merged.parquet', index=False)
```
